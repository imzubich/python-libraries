```python
import pandas as pd
import matplotlib.pyplot as plt
file = pd.read_csv("sales_data_sample.csv", encoding='latin1')
# print(file.columns)
x = file['ORDERDATE'][:50]
y = file['SALES'][:50]
plt.xticks(rotation=70)
color = ['red' if y<2500  else 'green' if y<3000 else 'black' if y<4000 else 'blue' for y in y]
# plt.xlabel("Sales")
# plt.ylabel("Date of Order")
plt.bar(x,y,color= color)
plt.show()
```
