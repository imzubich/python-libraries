```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.read_csv("house_data.csv")
data.columns
features = ["bedrooms","bathrooms","floors","view"]
target = "price"

X = data[features]
y = data[target]
X_train, X_test, y_train, y_test = train_test_split(X,y)

model = LinearRegression()
model.fit(X_train,y_train)

new_house = [[8, 15, 1, 1]]
predicted_price = model.predict(new_house)
print(f"Predicted price for the new house: $ {predicted_price[0]:.0f}")

print("\n____________PLots____________\n\n")


plt.scatter(data['bedrooms'], data['price'], color='blue', label='Actual Data')
plt.scatter(data['view'], data['price'], color='g', label='Actual Data',alpha = 0.1)
# plt.scatter(data['floors'], data['price'], color='blue', label='Actual Data')

# Also plot the predicted point
plt.scatter(new_house[0][0], predicted_price, color='red', label='Predicted Point', marker='D', s=150)

plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.title('Bedrooms vs Price')
plt.legend()
plt.show()
```
