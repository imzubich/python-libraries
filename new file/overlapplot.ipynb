```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = ["ML","DL","AI","NLP"]
y = [85,87,70,90]
z = [40,45,30,50]
a = [60,20,80,75]
b = [95,25,87,85]
width = 0.2
p = np.arange(len(x))
# p1 = p+width
# p2 = p + width*2
p1 = [j+width for j in p]
p2 = [i+width*2 for i in p]
p3 = [l+width*3 for l in p]
# print("values of p",p)

plt.title("Percentage Graph",fontsize = 30, fontweight = "bold")
plt.xlabel("Languages")
plt.ylabel("Values")
plt.xticks(ticks = p+0.3,rotation = 30)


plt.bar(x,y,width,color='b')
plt.bar(p1,z,width,color='r')
plt.bar(p2,a,width,color='g')
plt.bar(p3,b,width,color='black')

plt.legend(["Pakistan","India","Sri Lanka","Nebia"])
```
